### 【Pythonの「真」と「偽」】

・値が空（''）、0、None、[]　⇒ 偽 (False) とみなす
・値が入っている（'a'）、スペース ' '、[1] ⇒ 真 (True) とみなす

【短絡評価】
and: 一つでも偽があれば偽。最初に偽になった時点の値を返しそこで評価は終了する。全て真の場合は最後の評価を返す。
or: 一つでも真があれば真。最初に真になった時点の値を返しそこで評価は終了する。全て偽の場合は最後の評価を返す。

```python
n = -9
n >= -9 and n <= 9
```

```python
# 整数値の桁数（ゼロ / 1桁 / 2桁以上）を判定
n = int(input("整数値："))
if n == 0:
    print("その値はゼロです")
elif n >= -9 and n <= 9:
    print("その値は1桁です")
else:
    print("その値は2桁以上です")

'''
上記プログラムでnに-50を代入した場合、まずif節がFalseとなり、続くelif節で「n >= -9」が「False」になった時点でこの節の結果は「False」となり「n <= 9」の評価は省略。else節の、print("その値は2桁以上です") が実行される
'''
```

```python
# 整数値の桁数が2桁以上かどうかを判定
n = int(input("整数値："))
if n <= -10 or n >= 10:
    print("その値は2桁以上です")
else:
    print("その値は2桁未満です")

'''
上記プログラムでnに-50を代入した場合、
if節では「n <= -10」が「True」になった時点で評価が終了し、「n >= 10」の評価は省略される。if節の結果はTrueとなりprint("その値は2桁以上です") が実行される
'''
```

```python
# 文字列が空で and の場合
name1, name2, name3, name4 = '', 'suzuki', 'tanaka', 'sato'
selected_name = name1 and name2 and name3 and name4
print(selected_name)
```

```python
# 文字列にスペース(文字ありと見做される)で and の場合
name1, name2, name3, name4 = '　', 'suzuki', 'tanaka', 'sato'
selected_name = name1 and name2 and name3 and name4
print(selected_name)
#スペースも値があるとみなされ、すべて値が入っているので最後が返される
```

```python
# 文字列にスペース(文字ありと見做される)で or の場合 
name1, name2, name3, name4 = ' ', 'suzuki', 'tanaka', 'sato'
selected_name = name1 or name2 or name3 or name4
print(selected_name)
#目には見えないけど'　'のスペースを結果として返している
```

```python
# 文字列が空で or の場合
name1, name2, name3, name4 = '', 'suzuki', 'tanaka', 'sato'
selected_name = name1 or name2 or name3 or name4
print(selected_name)
# 最初に真になった値が返る
```

```python
'''
同一性判定演算子 x is y / x is not y と比較演算子 == / != の違い
==（等価性）：中身（値）が同じかどうかを判定
is（同一性）：実体（メモリ上の箱・id）が全く同じかどうかを判定。主に x is None （確実にNoneオブジェクトそのものか判定する時）などに使う。
'''

# 中身が全く同じ2つのリストを作る（双子）
list_a = [1, 2, 3]
list_b = [1, 2, 3]

# list_c には、list_a そのものを代入（別名をつけるだけ）
list_c = list_a

# ① == の判定（中身の勝負）
print(list_a == list_b)  # ⇒ True （中身は[1, 2, 3]で同じだから）

# ② is の判定（実体の勝負）
print(list_a is list_b)  # ⇒ False （別々のメモリに作られた別の箱だから）
print(list_a is list_c)  # ⇒ True （全く同じ箱を、aとcという2つの名前で呼んでいるだけだから）
```

```python
'''
Pythonのエコシステム

Pythonは、「-5 から 256 までのよく使う小さな整数」 や 「短い文字列」 については、「毎回新しい箱（メモリ）を作るのはもったいない」と考え、最初から用意してある1つの箱をみんなで使い回すという裏ルールを持っている。

逆に x = 257 と y = 257 だと、256を超えているので別々の箱が作られ、is は False になる。
'''

v = 100
w = 100
print(v is w)

x = 257
y = 257
print(x is y)
```
